```python
# Import dependencies
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\USER\OneDrive\Desktop\data-science-using-python\dirty_cafe_sales_cleaned.csv") 
df2 = df.copy()

df2.head()
```

```python
# Convert numeric columns properly
df2["Quantity"] = pd.to_numeric(df2["Quantity"], errors="coerce")
df2["Price Per Unit"] = pd.to_numeric(df2["Price Per Unit"], errors="coerce")
df2["Total Spent"] = pd.to_numeric(df2["Total Spent"], errors="coerce")

# Remove rows with missing numeric values
df2 = df2.dropna(subset=["Quantity", "Price Per Unit", "Total Spent"])

df2.head()
```

```python
plt.figure(figsize=(10,4), dpi=100)
sns.boxplot(x="Total Spent", data=df2)
plt.title("Outliers in Total Spent")
plt.show()
```

```python
plt.figure(figsize=(10,4), dpi=100)
color_palette = sns.color_palette("Accent_r")
sns.set_palette(color_palette)

sns.countplot(x="Item", data=df2)
plt.title("Count of Items Sold")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,4), dpi=100)
color_palette = sns.color_palette("cool")
sns.set_palette(color_palette)

sns.countplot(x="Payment Method", data=df2)
plt.title("Payment Method Distribution")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,4), dpi=100)
color_palette = sns.color_palette("magma")
sns.set_palette(color_palette)

sns.countplot(x="Location", data=df2)
plt.title("Sales by Location")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,4), dpi=100)
color_palette = sns.color_palette("plasma")
sns.set_palette(color_palette)

sns.barplot(x="Item", y="Total Spent", data=df2)
plt.title("Average Total Spent per Item")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,4), dpi=100)
color_palette = sns.color_palette("viridis")
sns.set_palette(color_palette)

sns.barplot(x="Location", y="Total Spent", data=df2)
plt.title("Location vs Total Sales")
plt.xticks(rotation=45)
plt.show()
```
